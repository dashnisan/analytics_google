#### **`device`**

_The type of device a user starts a session with_

This is a categorical variable, so you do not plot a box plot for it. A good plot for a binary categorical variable is a pie chart.

```python
# Pie chart
### YOUR CODE HERE ###

def plot_feature_pie(df, feature):
    uniques=df[feature].unique()
    s2list = list(df[feature])
    counts={item_type:s2list.count(item_type) for item_type in uniques}
    items = tuple(counts.items())
    keys, values = zip(*items)
    
    plt.pie(data=df, x=values, labels=keys, autopct='%.0f%%')
    plt.title('{} Proportion in Analyzed Month'.format(feature))
    #plt.xticks(rotation=45, horizontalalignment='right', ticks=range(0,22500,1750), labels=range(0,22500,1750))
    plt.show()
    
plot_feature_pie(df, 'device')
```

### **Task 3b. Handling outliers**

The box plots from the previous section indicated that many of these variables have outliers. These outliers do not seem to be data entry errors; they are present because of the right-skewed distributions.

Depending on what you'll be doing with this data, it may be useful to impute outlying data with more reasonable values. One way of performing this imputation is to set a threshold based on a percentile of the distribution.

To practice this technique, write a function that calculates the 95th percentile of a given column, then imputes values > the 95th percentile with the value at the 95th percentile.  such as the 95th percentile of the distribution.



```python
### YOUR CODE HERE ###

def quantile_replace_outliers(num_series, qtag):
    '''num_seres: a numerical pandas series
    qtag: the quantile to be calculated and used for replacement of outliers. As +real number <=1'''
    #print(num_series.max())
    qvalue = num_series.quantile(q=qtag)
    new_series = num_series.where(num_series<qvalue, qvalue)
    #mask = num_series > qvalue
    #num_series[mask] = qvalue
    #print(new_series.max())

    #return qvalue, num_series
    return qvalue, new_series
```

Next, apply that function to the following columns:
* `sessions`
* `drives`
* `total_sessions`
* `driven_km_drives`
* `duration_minutes_drives`

```python
### YOUR CODE HERE ###

df_fnooutl = df_filter.copy() # with no copy function, only an equal operator, weird things happen in memory!!!
# That way the new DF is just a reference of the original one -seems to!

columns = ['sessions', 'drives', 'total_sessions', 'driven_km_drives', 'duration_minutes_drives']

for column in columns:
    #print(df_filter[column].max())
    quantile, tempseries = quantile_replace_outliers(df_fnooutl[column], 0.95)
    #print(df_filter[column].max())
    print('{}:\n max original = {:.3e}\n max_nooutliers = {:.3e}\n'
          .format(column, df_fnooutl[column].max(), tempseries.max()))
    df_fnooutl[column] = tempseries
    #print(df_filter[column].max())
    #print('\n')

    #break
```
